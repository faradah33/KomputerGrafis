```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import io

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)', color='blue')
plt.title('Grafik Sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)

plt.savefig('grafik_sin.png', format='png')
img = Image.open('grafik_sin.png')
img.show()
```
